```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data/house_price.csv')
price_location = df.groupby('neighborhood').mean()
price_location.plot(kind='bar', title='Average Price per Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('./data/house_price.csv')

max_row = df.loc[df['sales_price'].idxmax()]

max_row_df = max_row.to_frame().T

max_row_df.to_csv('./data/highest_price_house.csv', index=False)

print("Exported to highest_price_house.csv")
```

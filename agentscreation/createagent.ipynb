```python
import ollama

# Simple trace function for labeling runs
def trace(label):
    print(f"\n--- TRACE: {label} ---")

class Agent:
    def __init__(self, name: str, instructions: str, model: str = "mistral"):
        self.name = name
        self.instructions = instructions
        self.model = model

    def run(self, prompt: str) -> str:
        response = ollama.chat(
            model=self.model,
            messages=[
                {"role": "system", "content": self.instructions},
                {"role": "user", "content": prompt}
            ],
        )
        return response["message"]["content"]

class Runner:
    @staticmethod
    def run(agent: Agent, prompt: str):
        class Result:
            def __init__(self, output):
                self.final_output = output
        output = agent.run(prompt)
        return Result(output)

# Create agents
joke_teller = Agent(
    name="Joke Teller",
    instructions="You are a funny joke teller. Always respond with a joke."
)

motivational_coach = Agent(
    name="Motivational Coach",
    instructions="You are a motivational coach. Always inspire and encourage."
)

math_tutor = Agent(
    name="Math Tutor",
    instructions="You are a patient math tutor. Explain answers clearly and simply."
)

# Prompts for each agent
prompts_jokes = [
    "Tell me a joke about relatives.",
    "Why am i feeling dizzy after programming"
]

prompts_motivation = [
    "I failed my test, what should I do?",
    "my relative told me im boycotted without telling exact reason"
]

prompts_math = [
    "Explain what a derivative is.",
    "What is 12 multiplied by 15?"
]

# Run and print results for each agent and prompts
for prompt in prompts_jokes:
    trace(f"Joke Teller prompt: {prompt}")
    result = Runner.run(joke_teller, prompt)
    print(result.final_output)

for prompt in prompts_motivation:
    trace(f"Motivational Coach prompt: {prompt}")
    result = Runner.run(motivational_coach, prompt)
    print(result.final_output)

for prompt in prompts_math:
    trace(f"Math Tutor prompt: {prompt}")
    result = Runner.run(math_tutor, prompt)
    print(result.final_output)
```
